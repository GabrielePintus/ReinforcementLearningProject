```python
import gymnasium as gym 
import numpy as np
from agent import SACAgent
from gymnasium import wrappers
import matplotlib.pyplot as plt
```

```python
agent = SACAgent(
    env = gym.make('BipedalWalker-v3', max_episode_steps=1600),
    discount_factor = 0.99,
    lr_actor = 0.0003,
    lr_critic_value = 0.0003,
    tau = 0.001,
    max_size= 4000000,
    batch_size = 256,
    reward_scale = 2,
    save_weights = True
)

rwd, avg_rwd, ep_steps = agent.learn(n_episodes=600)
```

```python
# Plotting the list
plt.plot(rwd)
plt.title('Episode Rewards')
plt.xlabel('Episode')
plt.ylabel('Total Reward')
plt.show()
```

```python
plt.plot(avg_rwd)
plt.title('Average Reward')
plt.xlabel('Episode')
plt.ylabel('Average Reward')
plt.show()
```

```python
plt.plot(ep_steps)
plt.title('Episode Steps')
plt.xlabel('Episode')
plt.ylabel('Steps')
plt.show()
```
